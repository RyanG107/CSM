```python
import numpy as np
import matplotlib.pyplot as plt
x=2
h=1e-10
print("estimate:",((x+h)**3-x**3)/h)
print("True:",3*x**2)
```

```python
h=1e-10
x=1
print("estimate:",(1/(x+h)-1/x)/h)
print("true",-1/x**2)
```

```python
import math
h=1e-10
x=3
print("estimate:",(math.sqrt(x+h)-math.sqrt(x))/h)
print("true",0.5/math.sqrt(x))
```

```python
def my_fxn(newx):
    newy=(newx**2-1)/(newx-1)
    return newy
my_fxn(2)
```

```python
x=np.linspace(-10,10,10000)
y=my_fxn(x)
```

```python
fig,ax=plt.subplots()
plt.axvline(x=0,color='blue')
plt.axhline(y=0,color='blue')
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.axvline(x=1,color='purple',linestyle='--')
plt.axhline(y=2,color='purple',linestyle='--')
_=ax.plot(x,y)
```

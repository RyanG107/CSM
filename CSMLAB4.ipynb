```python
import numpy as np
import matplotlib.pyplot as plt
x=2
h=1e-10
print("estimate:",((x+h)**3-x**3)/h)
print("True:",3*x**2)
```

```python
import math
h=1e-10
x=3
print("estimate:",(math.sqrt(x+h)-math.sqrt(x))/h)
print("true",0.5/math.sqrt(x))
```
